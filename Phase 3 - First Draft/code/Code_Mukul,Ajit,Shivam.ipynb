```python
import numpy as np
import math
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Cross Section type data of 32 States and UTs of India is taken for the year 2011-12. 
# The dependent variable used is the Poverty Rate (pov) in each state. 
# The primary independent variable used is the Gross Enrolment Ratio (ger) for classes IX-XII. 
# The other independent variables are Literacy Rate, Labour Force Participation Rate, Unemployment Rate, Per Capita Net State Domestic Product and Percentage of Population in Urban Area.
```

## Downloading and reading data

```python
# getting companies data
df = pd.read_csv("data.csv")
print(df.shape)
df
```

# Regression Models

## Simple Regression Model
### Model 1

$ pov =\beta_{0}+\beta_{1}\big(ger\big)\pm\varepsilon $


```python
import statsmodels.api as sm

# y : the variable which is dependent on x
# x : the independent variable

X = df['ger']
Y = df['pov']

X = sm.add_constant(X) # # adding the constant term


# performing the regression and fitting the model
model = sm.OLS(Y, X).fit()

# printing the summary table
summary_table = model.summary()
print(summary_table)
```

## Multiple regression models
### Model 2
$ pov =\beta_{0}+\beta_{1}\big(ger\big)+\beta_{2}\big(lit\big)+\beta_{3}\big(lfpr\big)+\beta_{4}\big(unemp\big)+\beta_{5}\big(nsdp\big)+\beta_{6}\big(urb\big)\pm\varepsilon $

```python
import statsmodels.api as sm

# y : the variable which is dependent on x
# x : the independent variable

X = df[['ger','lit','lfpr','unemp','nsdp','urb']]
Y = df['pov']

X = sm.add_constant(X) # # adding the constant term


# performing the regression and fitting the model
model = sm.OLS(Y, X).fit()

# printing the summary table
summary_table = model.summary()
print(summary_table)
```

### Model 3
$ pov =\beta_{0}+\beta_{1}\big(ger\big)+\beta_{2}\big(lfpr\big)+\beta_{3}\big(nsdp\big)\pm\varepsilon $

```python
import statsmodels.api as sm

# y : the variable which is dependent on x
# x : the independent variable

X = df[['ger','lfpr','nsdp']]
Y = df['pov']

X = sm.add_constant(X) # # adding the constant term


# performing the regression and fitting the model
model = sm.OLS(Y, X).fit()

# printing the summary table
summary_table = model.summary()
print(summary_table)
```
